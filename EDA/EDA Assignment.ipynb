### Read in the Datasets

To begin our EDA, we first need to read in both `performance_math.csv` and `performance_lang.csv`.

```python
Math = pd.read_csv('performance_math.csv')
print(Math.info())
Math.head(3)
```

```python
Lang = pd.read_csv('performance_lang.csv')
print(Lang.info())
Lang.head(3)
```

To ease the observation of the grades of a students, I will create a new column called **Average** to find the `Average Grade` of the students in the 3 Terms.

After that, I will divide them into 3 groups, namely **Low**, **Average** and **High** according to their Average Grades. The function is as below:

```python
def FindAverage(dataframe):
    dataframe['Average'] = (dataframe['T1'] + dataframe['T2'] + dataframe['T3'])/3.0
```

```python
def dis(Average):
    if(Average < 7):
        return('Low')
    elif(Average >= 7 and Average < 14):
        return('Average')
    elif(Average >= 14):
        return('High')
```

```python
FindAverage(Math)
Math['Grade'] = Math['Average'].apply(dis)
Math.head(3)
```

Here, we plot a countplot to see the overall distribution of students' **Average Grade**.

```python
plt.figure(figsize = (8,8))
    
sb.countplot(Math['Grade'], order = ['Low', 'Average', 'High'], palette = 'Set1')
    
plt.title('Grade Distribution for Math for both Schools', fontsize = 12)
plt.xlabel('Grade', fontsize = 12)
plt.ylabel('Count', fontsize = 12)
```

```python
# Write function to plot graphs, taking column name and dataframe as parameter
def grp_box(cat, dataframe):    
    sb.boxplot(x = cat, y = 'Average', data = dataframe, ax = axes[0])
    
    sb.heatmap(dataframe.groupby([cat, 'Grade']).size().unstack(), 
               linewidths = 1, annot = True, fmt = 'g', 
               annot_kws = {"size": 18}, cmap = "Blues", ax=axes[1])
```
